```python
import pandas as pd
import geocoder
```

#### Import the dataframe from Part I

```python
df_ca_postal = pd.read_csv(r'toronto_postal.csv')
df_ca_postal.head(10)
```

#### Get geo location (latitude, longitude) for each of the postal code in our dataframe

The code first creates two empty list (one for latitude, one for logitude)
It will then run a for loop for each of the postal code in our dataframe and retrieve the latitude and longitude of the corresponding postal code. Once retrieved, it will append the respective list. The code then create a new column in the dataframe, Latitude, Longitude, where it takes the value of the newly appended list.

```python
list_latitude = []
list_longitude = []
```

```python
for postal_code in df_ca_postal['Postcode']:
    
    lat_lng_coords = None
    
    while(lat_lng_coords is None):
        g = geocoder.google('{}, Toronto, Ontario'.format(postal_code))
        lat_lng_coords = g.latlng
        
    list_latitude.append(lat_lng_coords[0])
    list_longitude.append(lat_lng_coords[1])
```

```python
df_ca_postal["Latitude"] = list_latitude
df_ca_postal["Longitude"] = list_longitude
```

#### Unfortunately, I was not able to get any return. The free google service was not available. Below, the code will download the geospatial data and match with our original dataframe

```python
df_geo = pd.read_csv('http://cocl.us/Geospatial_data')
df_geo.head(10)
```

Now rename the column name to 'Postcode' to match the original dataframe so that we can merge the two dataframe on 'Postcode'

```python
df_geo.rename(columns={'Postal Code':'Postcode'}, inplace = True)
df_geo.head(10)
```

```python
df_ca_postal_geo = pd.merge(df_ca_postal, df_geo, on=['Postcode'], how = 'inner')
df_ca_postal_geo.head(10)
```

#### Export the dataframe to csv in order to import it for Part III

```python
df_ca_postal_geo.to_csv(r'toronto_postal_geo.csv', index = False)
```
