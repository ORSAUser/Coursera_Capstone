### Part I

```python
import pandas as pd
```

#### Read the table from wikipedia and make it into dataframe

This uses pandas read_html. It will read in all the tables from the page.
The first table (as annotated by [0]) is the table we're looking for.

```python
ca_postal = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
df_ca_postal = ca_postal[0]
df_ca_postal.head(10)
```

#### Drop all rows that have value 'Not assigned' for 'Borough'

The code first creates the index for all the Borough that has the value 'Not assigned'.
It then deletes the row with the identified index.

```python
indexNames = df_ca_postal[df_ca_postal['Borough'] == 'Not assigned'].index
df_ca_postal.drop(indexNames, inplace = True)
df_ca_postal.head(10)
```

#### Combine Neighborhood names for duplicate postal code

The first groups the dataframe by 'Postcode', then returns the first value from Borough for column 'Borough' and joins the Neighborhood values for column 'Neighborhood'

```python
df_ca_postal = df_ca_postal.groupby('Postcode').agg({'Borough':'first',
                                                    'Neighborhood':', '.join}).reset_index()
df_ca_postal.head(10)
```

#### Change Neighborhood name that is 'Not assigned' to the Borough name

The code iterates through the dataframe, when it sees a value 'Not assigned' in Neighborhood, it will replace it with the corresponding value in 'Borough'.

```python
for key, value in df_ca_postal.iterrows():
    if value.Neighborhood == 'Not assigned':
        value.Neighborhood = value.Borough
df_ca_postal.head(10)
```

#### Export the dataframe to a csv file in order to be import it for Part II

```python
df_ca_postal.to_csv(r'toronto_postal.csv', index = False)
```
